## Dostęp do wartości

```
moja_lista = ["Python", "jest", "nadal", "super"]
print(moja_lista[0])
print(moja_lista[3])
```

```
wspolrzedne = [[12.0, 13.3], [0.6, 18.0], [88.0, 1.1]]  # lista dwuwymiarowa
print(f"pierwsza współrzędna: {wspolrzedne[0]}")
print(f"drugi element pierwszej współrzędnej: {wspolrzedne[0][1]}")
```

## Aktualizowanie wartości

```
moja_lista = [0, 1, 2, 3, 4, 5]
moja_lista[0] = 99
print(moja_lista)

# usuń pierwszą wartość (o indeksie równym 0)
del moja_lista[0]
print(moja_lista)
```

## Sprawdzanie, czy dana wartość jest obecna na liście

```
jezyki = ["Java", "C++", "Go", "Python", "JavaScript"]
if "Python" in jezyki:
    print("Python jest na liście!")
```

```
if 6 not in [1, 2, 3, 7]:
    print("liczba 6 nie jest obecna")
```

## Listy są mutowalne

```
oryginalna = [1, 2, 3]
zmodyfikowana = oryginalna
zmodyfikowana[0] = 99
print(f"oryginalna: {oryginalna}, zmodyfikowana: {zmodyfikowana}")
```

Jak poprawnie tworzyć kopię listy?

```
oryginalna = [1, 2, 3]
zmodyfikowana = oryginalna.copy()
# Alternatywnie możesz użyć metody list
# zmodyfikowana = list(oryginalna)
zmodyfikowana[0] = 99
print(f"oryginalna: {oryginalna}, zmodyfikowana: {zmodyfikowana}")
```

## `list.append()`

```
moja_lista = [1]
moja_lista.append("coś")
print(moja_lista)
```

## `list.remove()`

```
moja_lista = ["Python", "jest", "czasami", "fajny"]
moja_lista.remove("czasami")
print(moja_lista)

# Jeśli nie jesteś pewien, czy wartość jest na liście, lepiej najpierw sprawdzić:
if "Java" in moja_lista:
    moja_lista.remove("Java")
else:
    print("Java nie jest częścią tej historii.")
```

## `list.sort()`

```
liczby = [8, 1, 6, 5, 10]
liczby.sort()
print(f"liczby: {liczby}")

liczby.sort(reverse=True)
print(f"liczby odwrotnie: {liczby}")

slowa = ["to", "jest", "lista", "slow"]
slowa.sort()
print(f"slowa: {slowa}")
```

## `sorted(list)`
Podczas gdy `list.sort()` sortuje listę w miejscu, `sorted(list)` zwraca nową listę i pozostawia oryginalną nietkniętą:

```
liczby = [8, 1, 6, 5, 10]
posortowane_liczby = sorted(liczby)
print(f"{liczby=}, {posortowane_liczby=}")
```

## `list.extend()`

```
pierwsza_lista = ["wolowina", "szynka"]
druga_lista = ["ziemniaki", 1, 3]
pierwsza_lista.extend(druga_lista)
print(f"{pierwsza_lista=}, {druga_lista=}")
```

Alternatywnie można również rozszerzać listy, sumując je:

```
pierwsza = [1, 2, 3]
druga = [4, 5]
pierwsza += druga  # to samo co: pierwsza = pierwsza + druga
print(f"{pierwsza=}")
```

## `list.reverse()`

```
moja_lista = ["a", "b", "szynka"]
moja_lista.reverse()
print(moja_lista)
```
