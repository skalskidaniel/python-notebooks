## `int`

```python
my_int = 6
print(f"value: {my_int}, type: {type(my_int)}")
```

## `float`

```python
my_float = float(my_int)
print(f"value: {my_float}, type: {type(my_float)}")
```

Zauważ, że dzielenie `int`ów daje w wyniku `float`:

```python
print(1 / 1)
print(6 / 5)
```

Uważaj na pułapki binarnego zapisu liczb zmiennoprzecinkowych (zobacz [Decimal](#decimal) jako obejście):

```python
val = 0.1 + 0.1 + 0.1
print(val == 0.3)
print(val)
```

## [`decimal.Decimal`](https://docs.python.org/3/library/decimal.html)

```python
from decimal import Decimal
```

```python
from_float = Decimal(0.1)
from_str = Decimal("0.1")
print(f"from float: {from_float}\nfrom string: {from_str}")
```

```python
my_decimal = Decimal("0.1")
sum_of_decimals = my_decimal + my_decimal + my_decimal
print(sum_of_decimals == Decimal("0.3"))
```

## Kolejność operatorów w obliczeniach
Obowiązuje matematyczna kolejność operatorów. Użyj nawiasów, jeśli chcesz zmienić kolejność wykonywania:

```python
print(1 + 2**2 * 3 / 6)  # 1 + 4 * 3 / 6 == 1 + 12 / 6 == 1 + 2
print((1 + 2**2) * 3 / 6)
```
