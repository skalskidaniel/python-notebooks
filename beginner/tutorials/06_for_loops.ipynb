# Pętle [`for`](https://docs.python.org/3/tutorial/controlflow.html#for-statements)

## Iterowanie po listach

```python
my_list = [1, 2, 3, 4, "Python", "is", "neat"]
for item in my_list:
    print(item)
```

### `break`
Zatrzymaj wykonywanie pętli.

```python
for item in my_list:
    if item == "Python":
        break
    print(item)
```

### `continue`
Przejdź do następnego elementu bez wykonywania linii występujących po `continue` wewnątrz pętli.

```python
for item in my_list:
    if item == 1:
        continue
    print(item)
```

### `enumerate()`
W przypadku, gdy potrzebujesz również znać indeks:

```python
for idx, val in enumerate(my_list):
    print(f"idx: {idx}, value: {val}")
```

## Iterowanie po słownikach

```python
my_dict = {"hacker": True, "age": 72, "name": "John Doe"}
for val in my_dict:
    print(val)
```

```python
for key, val in my_dict.items():
    print(f"{key}={val}")
```

## `range()`

```python
for number in range(5):
    print(number)
```

```python
for number in range(2, 5):
    print(number)
```

```python
for number in range(0, 10, 2): # ostatni to krok
    print(number)
```
