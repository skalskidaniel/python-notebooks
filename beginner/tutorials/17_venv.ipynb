# Wirtualne środowisko
Podczas pracy z projektami w Pythonie, najlepszą praktyką jest posiadanie oddzielnego wirtualnego środowiska dla każdego z nich.

Każde wirtualne środowisko ma swój własny plik binarny Pythona. Kiedy instalujesz jakiś pakiet Pythona w swoim wirtualnym środowisku, zostanie on zainstalowany tylko w tym konkretnym środowisku. Oznacza to, że możesz mieć różne wersje jednego pakietu Pythona w różnych wirtualnych środowiskach na tej samej maszynie. Wirtualne środowiska są również przydatne, jeśli potrzebujesz używać różnych wersji Pythona w swoich projektach.

## [`venv`](https://docs.python.org/3/library/venv.html#module-venv)

#### Tworzenie nowego wirtualnego środowiska
Możesz tworzyć wszystkie swoje wirtualne środowiska w jednym katalogu (na przykład, katalog `.virtualenvs` w twoim folderze domowym). Ułatwia to ich odnalezienie.

`python3 -m venv /path/to/new/environment`

lub

`path/to/your/python -m venv /path/to/new/environment`

#### Aktywacja wirtualnego środowiska

Windows: `path_to_virtual_env\Scripts\activate.bat`<br/>
Posix: `source path_to_virtual_env/bin/activate`

#### Instalowanie pakietów
Po aktywacji nowo utworzonego wirtualnego środowiska, możesz instalować nowe pakiety za pomocą `pip`. Na przykład, jeśli chcesz zainstalować `pytest`:

`python -m pip install pytest`

zostanie on zainstalowany w path_to_virtual_env/lib/<python_version>/site-packages. Zauważ, że ścieżka do site-packages może się nieznacznie różnić w zależności od używanego systemu operacyjnego.

Możesz wylistować zainstalowane pakiety i ich wersje, uruchamiając:

`python -m pip freeze`
