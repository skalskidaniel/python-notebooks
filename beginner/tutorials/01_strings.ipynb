## `str.replace()`

Metoda nie zmodyfikuje `my_string`, ponieważ zamiana nie jest wykonywana w miejscu.

```python
moj_string.replace("a", "?")
print(moj_string)
```

Zamiast tego musimy zapisać wartość zwracaną przez `replace`.

```python
my_modified_string = moj_string.replace("Python", "C++")
print(my_modified_string)
```

## `str.join()`

```python
pandas = "pandas"
numpy = "numpy"
requests = "requests"
fajne_biblioteki = ", ".join([pandas, numpy, requests])
```

```python
print(f"Fajne biblioteki Pythona: {fajne_biblioteki}")
```

Alternatywnie (znacznie wolniej)

```python
fajne_biblioteki = pandas + ", " + numpy + ", " + requests
print(f"Fajne biblioteki Pythona: {fajne_biblioteki}")

fajne_biblioteki = pandas
fajne_biblioteki += ", " + numpy
fajne_biblioteki += ", " + requests
print(f"Fajne biblioteki Pythona: {fajne_biblioteki}")
```

## `str.strip()`

```python
brzydki_string = " \n \t Jakieś    zdanie   "
stripped = brzydki_string.strip()

print(f"Brzydko: {brzydki_string}")
print(f"Ładnie: {stripped}")
```

## `str.split()`

```python
zdanie = "trzy różne słowa"
slowa = zdanie.split()
print(slowa)
```

```python
type(slowa)
```

```python
dane_binarne = "01001,101101,11100000"
liczby_binarne = dane_binarne.split(",")
print(liczby_binarne)
```

## Wywoływanie wielu metod pod rząd

```python
brzydki_string = "   tO wYGląDa ŹlE "
print(brzydki_string)
ladny_string = brzydki_string.strip().lower().replace("źle", "dobrze")
print(ladny_string)
```

Zauważ, że kolejność wykonywania jest od lewej do prawej. Dlatego to nie zadziała:

```python
ladny_string = brzydki_string.replace("źle", "dobrze").strip().lower()
print(ladny_string)
```
