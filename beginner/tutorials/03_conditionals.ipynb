# Warunki

## Testowanie wartości prawdy

### `bool`

```
print(f"typ True i False: {type(True)}")
```

```
print(f"0: {bool(0)}, 1: {bool(1)}")
print(f"pusta lista: {bool([])}, lista z wartościami: {bool(['hura'])}")
print(f"pusty słownik: {bool({})}, słownik z wartościami: {bool({'Python': 'super'})}")
```

### `==, !=, >, <, >=, <=`

```
print(f"{1 == 0}")
print(f"{1 != 0}")
print(f"{1 > 0}")
print(f"{1 > 1}")
print(f"{1 < 0}")
print(f"{1 < 1}")
print(f"{1 >= 0}")
print(f"{1 >= 1}")
print(f"{1 <= 0}")
print(f"{1 <= 1}")
```

Możemy łączyć warunki:

```
print(f"{1 <= 2 <= 3}")
```

### `and, or, not`

```
python_jest_super = True
java_jest_super = False
pusta_lista = []
tajna_wartosc = 3.14
```

```
print(f"Python i java są super: {python_jest_super and java_jest_super}")
print(f"tajna_wartosc i python_jest_super: {tajna_wartosc and python_jest_super}")
```

```
print(f"Python lub java jest super: {python_jest_super or java_jest_super}")
print(f"{1 >= 1.1 or 2 < 1.4}")
```

```
print(f"Java nie jest super: {not java_jest_super}")
```

Możemy kontrolować kolejność wykonywania za pomocą nawiasów.

```
print(bool(not java_jest_super or tajna_wartosc and python_jest_super or pusta_lista))
print(bool(not (java_jest_super or tajna_wartosc and python_jest_super or pusta_lista)))
```

## `if`

```
stwierdzenie = True
if stwierdzenie:
    print("stwierdzenie jest prawdziwe")

if not stwierdzenie:
    print("stwierdzenie nie jest prawdziwe")
```

```
lista = []

if lista:
    print("pusta lista nie zostanie oceniona jako True")
```

```
wartosc = 3
if 0 <= wartosc < 1 or wartosc == 3:
    print("Wartość jest dodatnia i mniejsza niż jeden lub wartość wynosi trzy")
```

## `if-else`

```
moj_slownik = {}
if moj_slownik:
    print("jest coś w moim słowniku")
else:
    print("mój słownik jest pusty")
```

## `if-elif-else`

```
wartosc = 88
if wartosc >= 100:
    print("wartość jest równa lub większa niż 100")
elif wartosc > 10:
    print("wartość jest większa niż 10 ale mniejsza niż 100")
else:
    print("wartość jest równa lub mniejsza niż 10")
```

Możesz mieć tyle instrukcji `elif`, ile potrzebujesz. Dodatkowo, `else` na końcu nie jest obowiązkowe.

```
powitanie = "Witaj Pythoniście!"
jezyk = "Wloski"

if jezyk == "Szwedzki":
    powitanie = "Hejsan!"
elif jezyk == "Finski":
    powitanie = "Latua perkele!"
elif jezyk == "Hiszpanski":
    powitanie = "Hola!"
elif jezyk == "Niemiecki":
    powitanie = "Guten Tag!"

print(powitanie)
```
