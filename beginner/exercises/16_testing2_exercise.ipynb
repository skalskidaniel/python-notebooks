# 1. Dokończ przypadki testowe
Część testująca implementacji `ListaZadan` jest niekompletna. Uzupełnij `____` w testach.

```
class ZadanieNieZnalezione(Exception):
    pass


class ListaZadan:
    def __init__(self):
        self._do_zrobienia = {}
        self._wykonane = {}
        self._licznik_zadan = 1

    @property
    def zadania_do_zrobienia(self):
        return self._do_zrobienia

    @property
    def zadania_wykonane(self):
        return self._wykonane

    def dodaj(self, zadanie):
        self._do_zrobienia[self._licznik_zadan] = zadanie
        self._licznik_zadan += 1

    def ukoncz(self, numer):
        if numer not in self._do_zrobienia:
            raise ZadanieNieZnalezione(f"{numer} nie ma w zadaniach do zrobienia")

        zadanie = self._do_zrobienia.pop(numer)
        self._wykonane[numer] = zadanie

    def usun(self, numer):
        if numer not in self._do_zrobienia:
            raise ZadanieNieZnalezione(f"{numer} nie ma w zadaniach do zrobienia")

        del self._do_zrobienia[numer]
```
