# 1. `if-elif-else`
Uzupełnij brakujące elementy (`____`) w poniższym kodzie, tak aby wydruki miały sens.

```python
imie = "Jan Kowalski"
```

```python
if ____:
    print(f'Imię "{imie}" ma więcej niż 20 znaków')
    opis_dlugosci = "dlugie"
elif ____:
    print(f'Imię "{imie}" ma więcej niż 15 znaków')
    opis_dlugosci = "pol dlugie"
elif ____:
    print(f'Imię "{imie}" ma więcej niż 10 znaków')
    opis_dlugosci = "pol dlugie"
elif ____:
    print(f'Imię "{imie}" ma 8, 9 lub 10 znaków')
    opis_dlugosci = "pol krotkie"
else:
    print(f'Imię "{imie}" to krótkie imię')
    opis_dlugosci = "krotkie"
```

```python
assert opis_dlugosci == "pol krotkie"
```
