# 1. `if-elif-else`
Uzupełnij brakujące elementy (`____`) w poniższym kodzie, tak aby wydruki miały sens.

```python
name = "John Doe"
```

```python
if ____:
    print(f'Imię "{name}" ma więcej niż 20 znaków')
    length_description = "long"
elif ____:
    print(f'Imię "{name}" ma więcej niż 15 znaków')
    length_description = "semi long"
elif ____:
    print(f'Imię "{name}" ma więcej niż 10 znaków')
    length_description = "semi long"
elif ____:
    print(f'Imię "{name}" ma 8, 9 lub 10 znaków')
    length_description = "semi short"
else:
    print(f'Imię "{name}" to krótkie imię')
    length_description = "short"
```

```python
assert length_description == "semi short"
```
