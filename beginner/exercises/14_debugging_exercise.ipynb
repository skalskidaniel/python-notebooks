# 1. Identyfikowanie błędów w kodzie
Poniższa funkcja `stripped_reversed_lowercase` zawiera co najmniej jeden błąd. Możesz to zobaczyć, uruchamiając kod w komórce poniżej, który testuje funkcjonalność funkcji `stripped_reversed_lowercase`.

Ustaw śledzenie na początku `stripped_reversed_lowercase` i użyj debugera, aby rozwiązać błąd (błędy). Wykonuj kod linia po linii i drukuj zmienne używane w funkcji, aby zrozumieć, co jest nie tak.

```python
def stripped_reversed_lowercase(original):
    # Ustaw tutaj punkt przerwania i rozpocznij debugowanie
    # from IPython.core.debugger import Pdb; Pdb().set_trace()
    stripped = original.lstrip()
    reversed = " ".join(reversed(stripped))
    reversed.lower()
    return reversed
```

```python
# Sprawdźmy, czy to działa
original = " \n Original String "
result = stripped_reversed_lowercase(original)
assert result == "gnirts lanigiro"
```
