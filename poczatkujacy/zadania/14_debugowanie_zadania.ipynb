# 1. Identyfikowanie błędów w kodzie
Poniższa funkcja `odwrocony_bez_spacji_male_litery` zawiera co najmniej jeden błąd. Możesz to zobaczyć, uruchamiając kod w komórce poniżej, który testuje funkcjonalność funkcji `odwrocony_bez_spacji_male_litery`.

Ustaw śledzenie na początku `odwrocony_bez_spacji_male_litery` i użyj debugera, aby rozwiązać błąd (błędy). Wykonuj kod linia po linii i drukuj zmienne używane w funkcji, aby zrozumieć, co jest nie tak.

```
def odwrocony_bez_spacji_male_litery(oryginalny):
    # Ustaw tutaj punkt przerwania i rozpocznij debugowanie
    # from IPython.core.debugger import Pdb; Pdb().set_trace()
    bez_spacji = oryginalny.lstrip()
    odwrocony = " ".join(reversed(bez_spacji))
    odwrocony.lower()
    return odwrocony
```

```
# Sprawdźmy, czy to działa
oryginalny = " \n Oryginalny Napis "
wynik = odwrocony_bez_spacji_male_litery(oryginalny)
assert wynik == "sipan ynlanygiro"
```
