# 2. Dokończ klasę `ManipulatorCiagu`
Uzupełnij `____` i utwórz implementację dla `obciete_tytul()`.

```python
class ManipulatorCiagu:
    """Docstring ManipulatorCiagu"""
    
    kategoria = "Manipulator"
    
    def __init__(self, oryginalny):
        self.ciag = oryginalny
        
    def odwroc_slowa(self):
        slowa = self.ciag.split()
        self.ciag = ' '.join(reversed(slowa))
        
    def zrob_tytul(self):
        # Utwórz implementację dla tego
        self.ciag = self.ciag.title()
        
    def zwroc_zmanipulowany(self):
        return self.ciag
```

```python
assert ManipulatorCiagu.__doc__ == "Docstring ManipulatorCiagu"
assert ManipulatorCiagu.kategoria == "Manipulator"

manip_ciagu = ManipulatorCiagu("fAjNy pyThON")

manip_ciagu.odwroc_slowa()
assert manip_ciagu.zwroc_zmanipulowany() == "pyThON fAjNy"

manip_ciagu.zrob_tytul()
assert manip_ciagu.zwroc_zmanipulowany() == "Python Fajny"
```
