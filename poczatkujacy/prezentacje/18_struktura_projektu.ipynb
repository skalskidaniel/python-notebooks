# Struktura projektu

## Skrypt Pythona
Python to świetny język do tworzenia małych narzędzi pomocniczych do różnego rodzaju zadań. Takie małe narzędzia często można wyrazić jako jednoplikowy skrypt Pythona.

Oto przykładowa struktura skryptu Pythona (czyli wykonywalnego modułu Pythona).

````python
# zawartość moj_skrypt.py

# importy
import logging

# stałe
LOGGER = logging.getLogger()


def magiczna_funkcja():
    LOGGER.warning("Zaraz zrobimy magiczne rzeczy")


def main():
    # Rzeczywista logika skryptu
    magiczna_funkcja()


if __name__ == "__main__":
    main()
````

## Pakiet Pythona
Przykładowa struktura projektu w Pythonie:

```
moj_projekt/
    README.md
    requirements.txt
    setup.py

    src/
        moj_projekt/
            __init__.py
            moj_modul.py
            inny_modul.py

            moj_pakiet1/
                __init__.py
                moj_trzeci_modul.py

    tests/
        conftest.py
        test_modul.py
        test_inny_modul.py

        moj_pakiet1/
            test_moj_trzeci_modul.py

```

* [requirements.txt](https://pip.pypa.io/en/latest/user_guide/#requirements-files) zawiera listę pakietów Pythona, od których zależy moj_projekt.
    * można je zainstalować, uruchamiając `pip install -r requirements`
* [setup.py](https://packaging.python.org/tutorials/distributing-packages/#setup-py) to plik, w którym umieszczasz istotne informacje o swoim projekcie, a plik ten jest również używany do pakowania projektu. Oto minimalny przykład pliku setup.py:

```python
'''Minimalny plik setup.py'''

from setuptools import setup, find_packages

setup(
    name='moj_projekt',
    version='0.1',
    packages=find_packages(where="src"),
    package_dir={"": "src"})
```
* Gdy masz już plik setup.py, możesz zainstalować swój projekt w trybie edytowalnym, uruchamiając `pip install -e .` w głównym katalogu projektu. W trybie edytowalnym zainstalowana wersja jest aktualizowana po wprowadzeniu zmian w plikach kodu źródłowego.
