### Argumenty

```python
def powitaj_nas(imie1, imie2):
    print(f"Witaj {imie1} i {imie2}!")


powitaj_nas("Jan Kowalski", "Superman")
```

```python
# Funkcja z wartością zwrotną
def oczysc_i_male_litery(oryginalny):
    zmodyfikowany = oryginalny.strip().lower()
    return zmodyfikowany


brzydki_string = "  MiEsZaNe LiTeRy "
ladny = oczysc_i_male_litery(brzydki_string)
print(f"ladny: {ladny}")
```

### Argumenty słów kluczowych

```python
def moje_wymyslne_obliczenie(pierwszy, drugi, trzeci):
    return pierwszy + drugi - trzeci


print(moje_wymyslne_obliczenie(3, 2, 1))

print(moje_wymyslne_obliczenie(pierwszy=3, drugi=2, trzeci=1))

# Z argumentami słów kluczowych możesz zmieniać kolejność
print(moje_wymyslne_obliczenie(trzeci=1, pierwszy=3, drugi=2))

# Możesz mieszać argumenty i argumenty słów kluczowych, ale musisz zacząć od argumentów
print(moje_wymyslne_obliczenie(3, trzeci=1, drugi=2))
```

### Domyślne argumenty

```python
def utworz_info_osoby(imie, wiek, praca=None, pensja=300):
    info = {"imie": imie, "wiek": wiek, "pensja": pensja}

    # Dodaj klucz 'praca' tylko jeśli jest podany jako parametr
    if praca:
        info.update(dict(praca=praca))

    return info


osoba1 = utworz_info_osoby("Jan Kowalski", 82)  # użyj domyślnych wartości dla praca i pensja
osoba2 = utworz_info_osoby("Anna Kowalska", 22, "haker", 10000)
print(osoba1)
print(osoba2)
```

**Nie używaj obiektów mutowalnych jako argumentów domyślnych!**

```python
def dodaj_jesli_wielokrotnosc_piec(liczba, magiczna_lista=[]):
    if liczba % 5 == 0:
        magiczna_lista.append(liczba)
    return magiczna_lista


print(dodaj_jesli_wielokrotnosc_piec(100))
print(dodaj_jesli_wielokrotnosc_piec(105))
print(dodaj_jesli_wielokrotnosc_piec(123))
print(dodaj_jesli_wielokrotnosc_piec(123, []))
print(dodaj_jesli_wielokrotnosc_piec(123))
```

Oto jak można osiągnąć pożądane zachowanie:

```python
def dodaj_jesli_wielokrotnosc_piec(liczba, magiczna_lista=None):
    if not magiczna_lista:
        magiczna_lista = []
    if liczba % 5 == 0:
        magiczna_lista.append(liczba)
    return magiczna_lista


print(dodaj_jesli_wielokrotnosc_piec(100))
print(dodaj_jesli_wielokrotnosc_piec(105))
print(dodaj_jesli_wielokrotnosc_piec(123))
print(dodaj_jesli_wielokrotnosc_piec(123, []))
print(dodaj_jesli_wielokrotnosc_piec(123))
```

### Docstringi
Ciągi znaków do dokumentowania funkcji, metod, modułów i zmiennych.

```python
def drukuj_sume(wartosc1, wartosc2):
    """Funkcja, która drukuje sumę podanych argumentów."""
    print(f"suma: {wartosc1 + wartosc2}")


print(help(drukuj_sume))
```

```python
def oblicz_sume(wartosc1, wartosc2):
    """To jest dłuższy docstring definiujący również argumenty i wartość zwrotną.

    Args:
        wartosc1: Pierwszy parametr.
        wartosc2: Drugi parametr.

    Returns:
        Suma wartosc1 i wartosc2.

    """
    return wartosc1 + wartosc2


print(help(oblicz_sume))
```
