## `int`

```python
my_int = 6
print(f"Wartość: {my_int}, typ: {type(my_int)}")
```

## `float`

```python
my_float = float(my_int)
print(f"Wartość: {my_float}, typ: {type(my_float)}")
```

Zauważ, że dzielenie `int`-ów daje w wyniku `float`:

```python
print(1 / 1)
print(6 / 5)
```

Uważaj na pułapki binarnego zapisu liczb zmiennoprzecinkowych (zobacz [Decimal](#decimal) jako obejście):

```python
val = 0.1 + 0.1 + 0.1
print(val == 0.3)
print(val)
```

## [`decimal.Decimal`](https://docs.python.org/3/library/decimal.html)

```python
from decimal import Decimal
```

```python
z_floata = Decimal(0.1)
ze_stringa = Decimal("0.1")
print(f"{z_floata = }\n{ze_stringa = }")
```

```python
moj_decimal = Decimal("0.1")
suma = moj_decimal + moj_decimal + moj_decimal
print(suma == Decimal("0.3"))
```

## Kolejność operatorów w obliczeniach
Obowiązuje matematyczna kolejność operatorów. Użyj nawiasów, jeśli chcesz zmienić kolejność wykonywania:

```python
print(1 + 2**2 * 3 / 6)  # 1 + 4 * 3 / 6 == 1 + 12 / 6 == 1 + 2
print((1 + 2**2) * 3 / 6)
```
