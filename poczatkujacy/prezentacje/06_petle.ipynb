# Pętle [`for`](https://docs.python.org/3/tutorial/controlflow.html#for-statements)

## Iterowanie po listach

```python
moja_lista = [1, 2, 3, 4, "Python", "jest", "super"]
for element in moja_lista:
    print(element)
```

### `break`
Zatrzymaj wykonywanie pętli.

```python
for element in moja_lista:
    if element == "Python":
        break
    print(element)
```

### `continue`
Przejdź do następnego elementu bez wykonywania linii występujących po `continue` wewnątrz pętli.

```python
for element in moja_lista:
    if element == 1:
        continue
    print(element)
```

### `enumerate()`
W przypadku, gdy potrzebujesz również znać indeks:

```python
for indeks, wartosc in enumerate(moja_lista):
    print(f"indeks: {indeks}, wartosc: {wartosc}")
```

## Iterowanie po słownikach

```python
moj_slownik = {"haker": True, "wiek": 72, "imie": "Jan Kowalski"}
for wartosc in moj_slownik:
    print(wartosc)
```

```python
for klucz, wartosc in moj_slownik.items():
    print(f"{klucz}={wartosc}")
```

## `range()`

```python
for liczba in range(5):
    print(liczba)
```

```python
for liczba in range(2, 5):
    print(liczba)
```

```python
for liczba in range(0, 10, 2): # ostatni to krok
    print(liczba)
```
